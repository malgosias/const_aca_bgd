## Goal ##

* Consider patching PEA to use a (low) constant background (in time).

* This is motivated by Obsid 18912 like events, when the estimated background gets too high due to warm/hot pixels and this makes the centroids and rates get into an unstable oscillation. See Tom's notebook at https://github.com/sot/skanb/blob/master/aca_track/simulate_track.ipynb.

* Use the ACA Level-0 flight data to evaluate the effect of underestimating the background (e.g. in the case of a high background anomaly) on star centroiding.

## Current on-board average background algorithm##

Implementation in https://github.com/sot/skanb/blob/master/aca_track/simulate_track.ipynb.

1. Compute the average signal from pixels A1, B1, G1, H1, I4, J4, O4, and P4.
2. Compare each of these pixel signals with the average. If any differ from the average by more than Â±150 percent or 10 A/D counts, whichever is larger, discard the one with the maximum deviation, or any one with the maximum deviation in case of a tie.
3. Repeat steps 1 and 2, with discarded pixels omitted, until no pixels are discarded at step 2. Then report the final result in the aspect telemetry data.

```python
def calc_raw_centroids_stats(d, slots, const_bgd=False):
    
    rows = []

    for slot in slots:
    
        slot_row = {'slot':slot, 'time': [],
                    'yan_raw':[], 'zan_raw':[]}

        imgs = aca_l0.get_l0_images(d.start, d.stop, slot=slot)
        
        # Calculate raw centroids
        for img in imgs:
            bgd = CONST_BGD if const_bgd else img.BGDAVG
            centroids = get_raw_centroids(img, bgd, img.IMGSIZE)
            
            # Translate (row, column) centroid to (yag, zag)
            y_pixels = centroids[1] + img.row0
            z_pixels = centroids[0] + img.col0
            
            try:
                yan, zan = transform.pixels_to_yagzag(y_pixels, z_pixels)
            except ValueError as e:
                print('{} at {}, slot {}'.format(e, DateTime(img.TIME).date, slot))
                print('(row0, col0) = ({}, {})'.format(img.row0, img.col0))
                print(img)
                yan, zan = (img.row0, img.col0)

            slot_row['yan_raw'].append(yan)
            slot_row['zan_raw'].append(zan)
            slot_row['time'].append(img.TIME)
                
        dt = -2.4    
    
        msids = ['aoacyan{}'.format(slot), 'aoaczan{}'.format(slot)]
    
        for msid in msids:
            telem_interp = interpolate(telem[msid].vals, telem[msid].times + dt,
                                       slot_row['time'], method="nearest")
            d_ags = slot_row['{}_raw'.format(msid[4:7])] - telem_interp
            resids = d_ags - np.median(d_ags)
            stds = np.std(resids)
            slot_row['{}_resids'.format(msid[4:7])] = resids
            slot_row['{}_stds'.format(msid[4:7])] = stds
                   
        rows.append(slot_row)
        
    return rows


def get_raw_centroids(raw_img, bgdavg, sz):

    if sz not in [6, 8]:
        centroids = [3.5, 3.5]
    else:
        raw_img = np.array(raw_img, dtype=float)        
        bgd_img = get_bgd_image(bgdavg, sz)    
        img = raw_img - bgd_img

        centered = False if sz == 6 else True
        img = zero_6x6_corners(img, centered=centered)
        if sz == 8:
            img = img[1:-1, 1:-1]
    
        pixels = np.arange(0, 6)
    
        # axis = 0 -- col, 1 -- row
        centroids = []
        for ax in [0, 1]:
            flat = np.sum(img, axis=ax)
            centroid = np.sum(flat * pixels) / np.sum(flat)
            if sz == 8:
                centroid += 1
            centroids.append(centroid)
    
    return centroids


def get_bgd_image(bgdavg, sz):
    """
    Return a (sz, sz) bgd array for bgd subtraction.
    ::bgdavg:: flight value of the average background
    """
    bgd_img = bgdavg * np.ones((sz, sz))
    return bgd_img


def zero_6x6_corners(img, centered=True):
    """
    ::img:: is a (6, 6) or (8, 8) array
    ::centered=True:: the 1:7 'corners' are set to zero,
                      relevant for the ER observations
    ::centered=False:: the 0:6 'corners' are set to zero,
                       relevant for the science observations
    """
    if np.shape(img) not in [(6, 6), (8, 8)]:
        raise TypeError("zero_6x6_corners:: Img should be a (6, 6) or (8, 8) array")
        
    if centered:
        r = (1, 1, 6, 6)
        c = (1, 6, 1, 6)
    else:
        r = (0, 0, 5, 5)
        c = (0, 5, 0, 5)
    for rr, cc in zip(r, c):
        img[rr, cc] = 0.0
    return img


def get_flight_bgd(dwell, slots):
    rows = []
    for slot in slots:
        row = {'time': [], 'bgdavg': []}
        imgs = aca_l0.get_l0_images(d.start, d.stop, slot=slot)
        for img in imgs:
            row['time'].append(img.TIME)
            row['bgdavg'].append(img.BGDAVG)
        rows.append(row)
    return rows


def plot_resids(t):
    fig = plt.figure(figsize=(10, 20))

    for slot in slots:
        ok = t['slot'] == slot
        for key in ['yan', 'zan']:
            plt.subplot(4, 2, slot + 1)
            times = t['time'][ok][0] - t['time'][ok][0][0]
            vals = t['{}_resids'.format(key)][ok][0]
            label = '{}, std = {:.3f}'.format(key, t['{}_stds'.format(key)][ok][0])
            plt.plot(times, vals, label=label)

        plt.ylim(-0.052, 0.052)
        plt.legend()
        plt.title("obsid {}, slot {}".format(obsid, slot))
        plt.ylabel("Resid (arcsec)")
        plt.xlabel("Time (sec)")

    plt.subplots_adjust(left=0.05, bottom=0.2, right=0.99, top=0.9, hspace=0.3, wspace=0.3)
    return


def plot_yanzan(t):
    fig = plt.figure(figsize=(10, 20))

    for slot in slots:
        ok = t['slot'] == slot
        for key in ['yan_raw', 'zan_raw']:
            plt.subplot(4, 2, slot + 1)
            times = t['time'][ok][0] - t['time'][ok][0][0]
            vals = t[key][ok][0] - t[key][ok][0][0]
            plt.plot(times, vals, label=key)
            msid = 'aoac{}{}'.format(key[:3], slot)
            times = telem[msid].times - telem[msid].times[0]
            vals = telem[msid].vals - telem[msid].vals[0]
            plt.plot(times, vals, label='{}_flight'.format(key[:3]))

        plt.legend()
        plt.title("Obsid {}, slot {}".format(obsid, slot))
        plt.ylabel("Resid (arcsec)")
        plt.xlabel("Time (sec)")

    plt.subplots_adjust(left=0.05, bottom=0.2, right=0.99, top=0.9, hspace=0.3, wspace=0.3)
    return


def plot_flight_bgd(rows):
    colors = 'rgbymckr'
    fig = plt.figure(figsize=(10, 15))

    for ii, row in enumerate(rows):
        plt.subplot(8, 1, ii + 1)
        plot_cxctime(row['time'], row['bgdavg'], color=colors[ii], label='Slot {}'.format(ii))
        plt.ylabel("Av. Flight Bgd (DN)")
        plt.margins(0.05)
        plt.title("Slot {}".format(ii))
    
    plt.subplots_adjust(left=0.05, bottom=0.2, right=0.99, top=0.9, hspace=0.3, wspace=0.3)
    return
```
